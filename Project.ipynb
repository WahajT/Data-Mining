```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix,classification_report

import numpy as np
```


# **Prediction before Mid-II**

1) Following is KNN model trained to
predict grade based on the first 4 assignments and quizzes, and the S-I marks. It is trained on the 80% of the data and tested on the remaining 20%


```python
# Select the first 4 assignment columns as feature_1, feature_2, feature_3, and feature_4
feature_1 = df['As:1'].values
feature_2 = df['As:2'].values
feature_3 = df['As:3'].values
feature_4 = df['As:4'].values

# Select the next 4 quiz columns as feature_5, feature_6, feature_7, and feature_8
feature_5 = df['Qz:1'].values
feature_6 = df['Qz:2'].values
feature_7 = df['Qz:3'].values
feature_8 = df['Qz:4'].values

# Select the mid-1 marks column as feature_9
feature_9 = df['S-I'].values

# Concatenate the feature arrays horizontally using numpy's hstack function
features = np.hstack((feature_1.reshape(-1, 1), feature_2.reshape(-1, 1), 
                      feature_3.reshape(-1, 1), feature_4.reshape(-1, 1), 
                      feature_5.reshape(-1, 1), feature_6.reshape(-1, 1), 
                      feature_7.reshape(-1, 1), feature_8.reshape(-1, 1),
                      feature_9.reshape(-1, 1)))
```

```python
target = df['Grade'].values
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
accuracy_knn = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy_knn:.2f}')

print("---------------------------------")
y_true = df['Grade'].tail(56)
tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("---------------------------------")


print("Classification Report:")
print(classification_report(y_test, y_pred))

print("---------------------------------")
sensitivity = tp / (tp + fn)
print("Sensitivity:", sensitivity)
```

```python
new_data = [[80, 60, 70, 25, 3, 11, 7.5, 8, 7.5]]
new_data_scaled = scaler.transform(new_data)
prediction = knn.predict(new_data_scaled)
print(f'Prediction: {prediction}')
```

2) Following is the Decision tree trained on the 80% of the total data and rest of the 20% is used for testing.

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)

dt_pred = dt.predict(X_test)
accuracy_dt = accuracy_score(y_test, dt_pred)
print("Accuracy : ",accuracy_dt)

print("---------------------------------")
y_true = df['Grade'].tail(56)
tn, fp, fn, tp = confusion_matrix(y_true, dt_pred).ravel()

print("Confusion Matrix:")
print(confusion_matrix(y_test, dt_pred))

print("---------------------------------")


print("Classification Report:")
print(classification_report(y_test, dt_pred))

print("---------------------------------")
sensitivity = tp / (tp + fn)
print("Sensitivity:", sensitivity)
```

```python
new_data = [[80, 60, 70, 25, 3, 11, 7.5, 8, 8.5]]
new_data_scaled = scaler.transform(new_data)
prediction_dt = dt.predict(new_data_scaled)
print(f'Prediction: {prediction_dt}')
```

3) Following is the Naive Bayes trained on the 80% of the data. Rest of 20% is used for testing.

```python
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb.fit(X_train, y_train)

nb_pred = nb.predict(X_test)
accuracy_nb = accuracy_score(y_test, nb_pred)
print(f'Accuracy: {accuracy_nb:.2f}')

print("---------------------------------")
y_true = df['Grade'].tail(56)
tn, fp, fn, tp = confusion_matrix(y_true, nb_pred).ravel()

print("Confusion Matrix:")
print(confusion_matrix(y_test, nb_pred))

print("---------------------------------")


print("Classification Report:")
print(classification_report(y_test, nb_pred))

print("---------------------------------")
sensitivity = tp / (tp + fn)
print("Sensitivity:", sensitivity)
```

```python
new_data = [[80, 60, 70, 25, 3, 11, 7.5, 8, 8.5]]
new_data_scaled = scaler.transform(new_data)
prediction_nb = nb.predict(new_data_scaled)
print(f'Prediction: {prediction_nb}')
```

**After Mid II**

A little preprocessing is needed before we start working on training model. We need to select the best 5 quizzes and assignments so first we have done that before training models to predict grade.

```python
df_best_as = df[['As:1', 'As:2', 'As:3', 'As:4', 'As:5', 'As:6', 'As:7']].apply(lambda row: row.nlargest(5).tolist(), axis=1, result_type='expand')
```

```python
df_best_as = df_best_as.rename(columns={0: 'As:1',1:'As:2',2:'As:3',3:'As:4',4:'As:5'})

df_best_as
```

```python
df_best_qz= df[['Qz:1', 'Qz:2', 'Qz:3', 'Qz:4', 'Qz:5', 'Qz:6', 'Qz:7','Qz:8']].apply(lambda row: row.nlargest(5).tolist(), axis=1, result_type='expand')
```
